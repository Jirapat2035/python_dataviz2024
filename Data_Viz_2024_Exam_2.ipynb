###สืบหาฆาตรกร ในคดี ฆาตรกรรม (murder) ที่เกิดขึ้นในวันที่ 15 มกราคม 2018 ที่เมือง SQL City โดยเริ่มจากการตรวจสอบ ข้อมูลคดี (crime_scene_report.csv)


```python
#สืบหาฆาตรกร ในคดี ฆาตรกรรม (murder) ที่เกิดขึ้นในวันที่ 15 มกราคม 2018 ที่เมือง SQL City โดยเริ่มจากการตรวจสอบ ข้อมูลคดี (crime_scene_report.csv)
df_crime_scene_report ['date'] = df_crime_scene_report ['date'].astype(str)
#แสดงผลการใช้ โดยใช้ตัวดำเนินการ 'และ' ระหว่าง 2 Conditons
result = df_crime_scene_report [(df_crime_scene_report ['date'].str.contains('20180115')) &
                (df_crime_scene_report ['type'] == 'murder') &
                (df_crime_scene_report ['city'] == 'SQL City')]
result
```
