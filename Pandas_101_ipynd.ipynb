##1.นิทานที่ยาวที่สุดและสั้นที่สุด (ไม่นับสัญลักษณ์พิเศษ)

```python
import re

longest_story_title = ""
shortest_story_title = ""
longest_length = 0
shortest_length = float('inf')

description_index = None
title_index = None

# หา index ของ column 'description' และ 'title'
for i, column_name in enumerate(folk_df.columns):
    if column_name == 'description':
        description_index = i
    elif column_name == 'title':
        title_index = i

# หานิทานที่ยาวที่สุดและสั้นที่สุด
for i in range(len(folk_df)):
    story = folk_df.iloc[i, description_index]
    if isinstance(story, str):
        # ใช้ regular expression เพื่อกรองเฉพาะตัวอักษรภาษาไทย
        thai_chars = re.findall(r'[\u0E00-\u0E7F]', story)
        story_length = len(thai_chars)
    else:
        story_length = 0

    if story_length > longest_length:
        longest_length = story_length
        longest_story_title = folk_df.iloc[i, title_index]

    if story_length < shortest_length:
        shortest_length = story_length
        shortest_story_title = folk_df.iloc[i, title_index]

print("นิทานที่ยาวที่สุด:", longest_story_title, "(จำนวนตัวอักษรไม่รวมเว้นวรรค:", longest_length, ")")
print("นิทานที่สั้นที่สุด:", shortest_story_title, "(จำนวนตัวอักษรไม่รวมเว้นวรรค:", shortest_length, ")")
```

##2.ชนิดของนิทาน (Type) ที่มากที่สุดและน้อยที่สุด

```python
# หาชนิดของนิทานที่ไม่ซ้ำกัน
unique_types = []
for i in range(len(folk_df)):
    story_type = folk_df.iloc[i, folk_df.columns.get_loc('type')]
    if story_type not in unique_types:
        unique_types.append(story_type)

# นับจำนวนชนิดของนิทานแต่ละชนิด
type_counts = {}
for type_name in unique_types:
    count = 0
    for i in range(len(folk_df)):
        if folk_df.iloc[i, folk_df.columns.get_loc('type')] == type_name:
            count += 1
    type_counts[type_name] = count

# หาชนิดของนิทานที่มากที่สุดและน้อยที่สุด
most_common_type = ""
least_common_type = ""
most_count = 0
least_count = float('inf')

for type_name, count in type_counts.items():
    if count > most_count:
        most_count = count
        most_common_type = type_name
if count < least_count:
        least_count = count
        least_common_type = type_name

print("ชนิดของนิทานที่มากที่สุด:", most_common_type, "(จำนวน:", most_count,")")
print("ชนิดของนิทานที่น้อยที่สุด:", least_common_type, "(จำนวน:", least_count,")")
```

##Quiz9.1
วนลูปนับว่ามีนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องกี่เรื่อง เรื่องอะไรบ้าง
มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน
คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุด

```python

# prompt: วนลูปนับว่ามีนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องกี่เรื่อง เรื่องอะไรบ้าง
# มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน เท่าไหร่
# คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุด

# นับจำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่อง
count_mom_dad = 0
mom_dad_stories = []
count_life_stories = 0
count_local_stories = 0
creator_counts = {} #กำหนด dictionary ให้ว่างๆก่อน

for _, row in this_data.iterrows():
    title = row['title']
    story_type = row['type']
    creator = row['creator']

    if 'แม่' in title or 'พ่อ' in title:
        count_mom_dad += 1
        mom_dad_stories.append(title)

        if story_type == 'นิทานชีวิต':
            count_life_stories += 1
        elif story_type == 'นิทานประจำถิ่น':
            count_local_stories += 1

        if creator in creator_counts:
            creator_counts[creator] += 1
        else:
            creator_counts[creator] = 1

# หาคนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า 'แม่' หรือ 'พ่อ' มากที่สุด
most_frequent_creator = max(creator_counts, key=creator_counts.get)

print("1.จำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่อง:", count_mom_dad)
print("- รายชื่อนิทาน:", mom_dad_stories)
print("2.1 จำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่อง และเป็นนิทานชีวิต:", count_life_stories)
print("2.2 จำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่อง และเป็นนิทานประจำถิ่น:", count_local_stories)
print("3. คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า 'แม่' หรือ 'พ่อ' มากที่สุด:", most_frequent_creator)
```
