####3.ตอบคำถามว่า data มีทั้งหมดกี่กลุ่ม โดย plot จุดข้อมูลแต่ละกลุ่มเป็นคนละสี


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# สมมติว่า df เป็น numpy array ขนาด (51, 65536)
num_samples, num_original_columns = df.shape

# 1. Preprocess ข้อมูล
# ตัดข้อมูลที่อยู่นอกช่วงที่สนใจ (9.5 ≤ ppm ≤ 0)
ppm = np.linspace(0, 10, num_original_columns)  # ตัวอย่างการสร้างค่า ppm
mask = (ppm >= 0) & (ppm <= 9.5)
filtered_df = df[:, mask]  # ตัดข้อมูลตาม mask

# 2. Quantize ข้อมูลโดยเลือกค่า max ในแต่ละกลุ่ม
num_quantized_columns = 400
group_size = filtered_df.shape[1] // num_quantized_columns

quantized_data = np.empty((num_samples, num_quantized_columns))

for i in range(num_quantized_columns):
    start_index = i * group_size
    end_index = start_index + group_size
    quantized_data[:, i] = np.max(filtered_df[:, start_index:end_index], axis=1)

# 3. ทำ PCA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(quantized_data)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# 4. ใช้ K-Means เพื่อจัดกลุ่มข้อมูล
num_clusters = 3  # สมมติว่าต้องการหารือกลุ่ม 3 กลุ่ม
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(pca_result)

# 5. แสดงผลลัพธ์
labels = kmeans.labels_

# กำหนดสีสำหรับแต่ละกลุ่ม
colors = ['red', 'green', 'blue']
group_colors = [colors[label] for label in labels]

# Visualize ข้อมูลแต่ละกลุ่มด้วยสีที่แตกต่างกัน
plt.figure(figsize=(10, 6))
for i in range(num_clusters):
    plt.scatter(pca_result[labels == i, 0], pca_result[labels == i, 1], alpha=0.7,
                color=colors[i], label=f'Cluster {i + 1}')

plt.title('K-Means Clustering - Visualization of Data Points by Cluster')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid()
plt.show()

# จำนวนกลุ่ม
print(f"จำนวนกลุ่มในข้อมูล: {num_clusters}")
```

####4.สร้าง ตารางที่มี column บอกว่า data แต่ละจุดนั้นอยู่กลุ่มไหน


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# สมมติว่า df เป็น numpy array ขนาด (51, 65536)
num_samples = 51
num_original_columns = 65536

# สร้างข้อมูลตัวอย่าง (แทนที่ด้วยข้อมูลจริงของคุณ)
np.random.seed(0)
df = np.random.rand(num_samples, num_original_columns)

# 1. Preprocess ข้อมูล
ppm = np.linspace(0, 10, num_original_columns)  # ตัวอย่างการสร้างค่า ppm
mask = (ppm >= 0) & (ppm <= 9.5)  # ตัดข้อมูลที่อยู่นอกช่วงที่สนใจ
filtered_df = df[:, mask]  # ตัดข้อมูลตาม mask

# 2. Quantize ข้อมูลโดยเลือกค่า max ในแต่ละกลุ่ม
num_quantized_columns = 400
group_size = filtered_df.shape[1] // num_quantized_columns

quantized_data = np.empty((num_samples, num_quantized_columns))

for i in range(num_quantized_columns):
    start_index = i * group_size
    end_index = start_index + group_size
    quantized_data[:, i] = np.max(filtered_df[:, start_index:end_index], axis=1)

# 3. ทำ PCA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(quantized_data)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# 4. ใช้ K-Means เพื่อจัดกลุ่มข้อมูล
num_clusters = 3  # กำหนดจำนวนกลุ่มที่ต้องการ
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(pca_result)

# 5. สร้าง DataFrame สำหรับข้อมูลที่ quantize และเพิ่มคอลัมน์กลุ่ม
quantized_df = pd.DataFrame(quantized_data, columns=[f'Feature_{i + 1}' for i in range(num_quantized_columns)])
quantized_df['Cluster'] = kmeans.labels_  # เพิ่มคอลัมน์กลุ่ม

# 6. เปลี่ยนตำแหน่งคอลัมน์ Cluster ให้เป็นคอลัมน์แรก
quantized_df = quantized_df[['Cluster'] + [f'Feature_{i + 1}' for i in range(num_quantized_columns)]]

# 7. แสดงผลตารางทั้งหมด 51 แถว
quantized_df  # แสดงข้อมูลทั้งหมด 51 แถว
```
